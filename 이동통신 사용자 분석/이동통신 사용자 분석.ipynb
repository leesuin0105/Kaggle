# 가입기간, 가입갱신여부, 고객센터 전화량 관계분석하기

```python
#가입기간, 가입갱신여부, 고객센터 전화량의 상관관계
df[['Leave','MemberWeeks', 'NewContract', 'Callcenter']].corr()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
corr = df[['Leave','MemberWeeks', 'NewContract', 'Callcenter']].corr()


mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True


sns.heatmap( corr,
 mask=mask,
 fmt='.2f',
 annot=True,
 ax=ax,
 cmap='RdBu',
 vmin=-1,
 vmax=1,
 square=True
)
```

1.
나의 예측 = 가입기간이 늘어날수록 가입갱신여부와 가입해지여부랑 강한 연관성이 있을것이다.

                결론 = 가입기간과 가입갱신여부는 약한 음의 상관관계 (-0.024735 값), 가입해지여부와는 약한양의상관관계(0.016541)를 가져 연관성이 거의 없다.

2.
나의 예측 = 가입갱신여부가 늘어날수록 고객센터 전화량은 많아져 강한 양의 상관관계를 가질것이다.

                결론 = 가입갱신여부와 고객센터 전화량은 약한 양의 상관관계 (0.024522 값)을 가지는것을 보아 연관성이 거의 없다.

# 월평균 요금분석

```python
#데이터사용량 기준으로 월평균요금금액 내림차순

(
df[['DataAmount','MonthlyFee','MemberWeeks']]
.groupby('DataAmount', as_index=True)
.apply(lambda df: df.sort_values('DataAmount', ascending=False).head(1))
.droplevel(0)
.sort_values(['DataAmount','MonthlyFee'], ascending=[False,False])
).head(10)
```

```python
#월평균전화량(분) 기준으로 월평균요금금액 내림차순
(
df[['CalMins','MonthlyFee','MemberWeeks']]
.groupby('CalMins', as_index=True)
.apply(lambda df: df.sort_values('CalMins', ascending=False).head(1))
.droplevel(0)
.sort_values(['CalMins','MonthlyFee'], ascending=[False,False])
).head(10)
```

데이터사용량이 높은 상위의 사람들을 보면 금액이 100.0000이상이 넘는 사람이 많고,

전화량이 높은 상위의 사람들을 보면 금액이 주로 70.0000 - 80.0000 정도대 인것을 보아

전화를 많이 사용하는것보다 데이터를 많이 사용하는것이 요금금액이 더 많이 나온다는 것을 알 수 있었다.

```python
# 조건검색 ( 3달~6달 사이 가입갱신하고 데이터 요금제 사용하는 사람의 예시)

c1 = (df.MemberWeeks >= 90) & (df.MemberWeeks <= 180)
c2 = df.NewContract == 1
c3 = df.IsData == 1

c_final = c1 & c2 & c3
```

```python
df[c_final].head(10)
```

```python
#월평균요금 수치형 -> 범주형 변환
#qcut: 분위수를 기준으로 나눔

(df.MonthlyFee.pipe(lambda ser: pd.qcut(ser, q=10)).value_counts().sort_index())
```
